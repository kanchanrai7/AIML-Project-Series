```python
import tkinter as tk
from tkinter import scrolledtext
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
from conversation_data import conversation

# Create a new chat bot named CollegeAdmissionBot
bot = ChatBot(
    'CollegeAdmissionBot',
    logic_adapters=[
        {
            'import_path': 'chatterbot.logic.BestMatch',
            'default_response': 'Please visit our website for the latest information on various deadlines',
            'maximum_similarity_threshold': 0.50
        }
    ]
)

# Create a new ListTrainer for the bot
trainer = ListTrainer(bot)



# Train the bot with the defined conversation
trainer.train(conversation)

# Create a Tkinter window
window = tk.Tk()
window.title("College Admission Inquiry Bot")

# Create a scrolled text widget for conversation display
conversation_display = scrolledtext.ScrolledText(window, width=80, height=20, wrap=tk.WORD)
conversation_display.grid(row=0, column=0, columnspan=2)

# Create an entry widget for user input
user_input_entry = tk.Entry(window, width=80)  # Increased width
user_input_entry.grid(row=1, column=0, padx=10, pady=10)

# Configure the "bot" tag for conversation display
conversation_display.tag_config("bot", foreground="blue")

def get_response():
    user_input = user_input_entry.get().lower()  # Convert to lowercase
    conversation_display.insert(tk.END, f"You: {user_input}\n", "user")  # User's text in default color
    
    if not user_input:
        # Handle empty input
        return
    
    if user_input in ['exit', 'bye', 'quit']:
        response = "Thank you for using CollegeAdmissionBot. Goodbye!"
    else:
        response = bot.get_response(user_input)
        
    conversation_display.insert(tk.END, f"Bot: {response}\n", "bot")  # Bot's text in blue
    
    user_input_entry.delete(0, tk.END)

# Create a button to get responses with increased size
ask_button = tk.Button(window, text="Ask", command=get_response, width=20, height=2)  # Increased size
ask_button.grid(row=1, column=1, padx=10, pady=10, sticky="e")

# Handle window closing event
def on_closing():
    # Perform cleanup actions if needed
    window.destroy()

window.protocol("WM_DELETE_WINDOW", on_closing)

# Start the Tkinter event loop
window.mainloop()
```
